```python
import librosa
import sounddevice as sd  
import numpy as np
import tensorflow as tf  
```

```python
model = tf.keras.models.load_model('tambhi.h5')
```

```python
def record_and_detect_weak_words():
    duration = 5  
    sample_rate = 8000  
    input_shape = (31, 16)  

    while True:
        print("Recording audio for {} seconds...".format(duration))
        audio_data = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1)
        sd.wait()  # Wait for recording to finish

        # Ensure the audio data is one-dimensional (monophonic)
        audio_data = audio_data.flatten()

        # Extract features from the audio data using Librosa
        mfccs = librosa.feature.mfcc(y=audio_data, sr=sample_rate, n_mfcc=input_shape[0])

        # Pad or truncate the feature matrix to match the expected input shape
        if mfccs.shape[1] < input_shape[1]:
            mfccs = np.pad(mfccs, ((0, 0), (0, input_shape[1] - mfccs.shape[1])), mode='constant')
        else:
            mfccs = mfccs[:, :input_shape[1]]

        # Perform weak word detection using your pre-trained model
        predictions = model.predict(np.expand_dims(mfccs, axis=0))

        threshold = 1  
        weak_words_detected = predictions > threshold

        if np.any(weak_words_detected):
            print("Weak words detected!")
        else:
            print("No weak words detected.")

if __name__ == "__main__":
    record_and_detect_weak_words()
```
